# Caminante aleatorio
Este consiste en lanzar una moneda a cada paso y tomar una decisi√≥n hacia donde caminar.

Lo que se hace es echar un volado y caminar hacia arriba o hacia abajo


```python
from pylab import *
%matplotlib inline
```

```python
trayectoria = zeros(1000)
posicion = 0 
for n in range(len(trayectoria)):
    trayectoria[n]=posicion
    volado = rand()
    if volado < 0.5:
        posicion = posicion - 1
    else:
        posicion = posicion + 1
plot(range(len(trayectoria)),trayectoria)
show()
```

```python
x=linspace(0,20)
y=(cos(3*x))/x
y2=1./x
plot(x,y,x,y2,x,-y2)
show()
```
